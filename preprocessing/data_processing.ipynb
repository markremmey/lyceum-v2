```python
import glob
text_files = glob.glob(os.path.join('text_files', '**/*.txt'), recursive=True)
```

## Split all text files into chunks. Save chunks to a JSON in memory

```python
import json
from langchain_text_splitters import CharacterTextSplitter
import tiktoken

book_list = []

model_name = 'gpt-3.5-turbo'

for i, file_path in enumerate(text_files):
  book_name = file_path.split('/')[-1].split('.')[0]
  book_list.append({
    'id': i,
    'title': book_name,
    'chunks': []
  })
  with open(file_path, 'r') as file:
    text = file.read()

  text_splitter = CharacterTextSplitter.from_tiktoken_encoder(
      chunk_size=200, chunk_overlap=20
  )
  texts = text_splitter.split_text(text)
  book_list[i]['chunks'] = texts
```

## Save JSON to a local directory

```python
print(book_list[0])

with open('../data/book_list.json', 'w') as file:
  json.dump(book_list, file)
```
